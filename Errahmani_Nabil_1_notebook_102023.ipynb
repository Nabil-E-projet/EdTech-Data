# Chapitre 2: Nettoyage des Données

## Aperçu
Le nettoyage des données est une étape fondamentale pour assurer la précision de l'analyse. Il s'agit de traiter les valeurs manquantes, ainsi que les doublons, supprimer les colonnes non pertinentes, et focaliser notre étude sur une période temporelle spécifique.

## Objectifs
- **Gérer les Valeurs Manquantes :** Identifier et traiter les valeurs manquantes pour améliorer la qualité du jeu de données.
- **Épurer le Jeu de Données :** Supprimer les colonnes superflues et celles avec un grand nombre de valeurs manquantes.
- **Focaliser sur la Période Pertinente :** Sélectionner les données entre 2000 et 2015, période la plus pertinente pour l'analyse d'expansion d'academy.
- **Suprresion des doublons :**  Supprimer les données qu'on retrouve au minimum 2 fois, doublons.

```python
# On configure pandas pour afficher toutes les lignes du dataframe.
pd.set_option('display.max_rows', None )

# On calcule et on affiche le pourcentage de valeurs manquantes pour chaque colonne dans dans le dataframe df (EdStatsData.csv).
df.isna().mean(axis=0)*100
```

On remarque que énormement de données sur la colonne années du fichier EdStatsData.csv sont manquant, et on a des colonnes Unnamed que j'estime inutile et qu'on peut supprimer. Quant aux années on devrai supprimer les années trop anciennes ou celle qui sont après notre époque, on gardera seulement pour notre analyse les années qui sont dans la période 2000 à 2015 (inclus).

```python
# On crée une liste des années en tant que chaînes de caractères( une liste ), de 2000 à 2015 inclus
years_list = [str(i) for i in range(2000, 2016)]  # on fait une boucle for qui avance d'année en année et en stock dans la liste

# On fait une liste avec les noms des informations qu'on veut garder, et on y ajoute les années pour lesquelles on veut aussi des informations.
selected_columns  = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'] + years_list

# On selectionnes uniquement les colonnes d'intérêt dans le dataframe EdStatsData.csv 
cleaned_df = df.loc[:, selected_columns]

# Enlever les lignes où il n'y a pas d'informations pour toutes les années choisies, pour ne garder que les données utiles.
cleaned_df = cleaned_df.dropna(how='all', subset=years_list)

# On affiche les informations du dataFrame nettoyé pour vérifier les résultats 
cleaned_df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Création d'une carte de chaleur pour visualiser les données manquantes
plt.figure(figsize=(10,6))
sns.heatmap(cleaned_df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Carte de chaleur des valeurs manquantes dans les données')
plt.show()
```

### Traitement des valeurs manquantes

Les valeurs manquantes dans les données peuvent biaiser notre analyse et conduire à des interprétations incorrectes. Pour y remédier, nous allons remplacer les valeurs manquantes par la médiane de chaque colonne correspondante. La médiane est choisie car elle est moins sensible aux valeurs aberrantes que la moyenne, ce qui peut être particulièrement utile dans les ensembles de données avec des variations importantes.


```python
for year in years_list:
    # Calcul de la moyenne globale pour l'année en cas de besoin
    global_mean = cleaned_df[year].mean()

    # Imputation conditionnelle avec une valeur de repli globale
    cleaned_df[year] = cleaned_df.groupby(['Indicator Name'])[year].transform(lambda x: x.fillna(x.mean() if not x.isna().all() else global_mean))

# Vérification des valeurs manquantes après l'imputation
missing_values_after = cleaned_df.isnull().sum().sum()
print(f"Nombre de valeurs manquantes après l'imputation conditionnelle avec la moyenne : {missing_values_after}")

```

Après avoir remplacé les valeurs manquantes par la médiane de chaque colonne, nous avons vérifié à nouveau la présence de valeurs manquantes dans le jeu de données. Le résultat montre qu'il n'y a plus de valeurs manquantes, ce qui signifie que notre jeu de données est maintenant complet et prêt pour une analyse plus approfondie.


### Analyse des données dupliquées

Après avoir traité les valeurs manquantes, il est également important de vérifier la présence de lignes dupliquées dans notre jeu de données. Les lignes dupliquées peuvent indiquer des redondances dans la collecte des données ou des erreurs lors de la consolidation des sources de données. Nous allons donc vérifier et compter le nombre de lignes dupliquées dans notre DataFrame nettoyé.


```python
# Vérification des données dupliquées
duplicates = cleaned_df.duplicated().sum()

if duplicates > 0:
    print(f"Le jeu de données contient {duplicates} lignes dupliquées.")
else:
    print("Pas de lignes dupliquées dans le jeu de données.")
```

L'absence de lignes dupliquées dans le jeu de données indique que notre ensemble de données est relativement propre à cet égard et ne souffre pas de redondances évidentes. Cela renforce la qualité de notre jeu de données pour les analyses ultérieures.

## Visualisation des Valeurs Aberrantes
Utilisation des box plots pour l'identification visuelle des valeurs aberrantes.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Liste des indicateurs à analyser
indicators = ['NY.GDP.PCAP.PP.KD', 'UIS.E.5.B', 'SE.TER.ENRR', 'UIS.E.3', 'SE.SEC.ENRR.UP', 'IT.NET.USER.P2']

# Création d'une figure pour les boxplots
plt.figure(figsize=(15, 10))

for i, indicator in enumerate(indicators, 1):
    plt.subplot(2, 3, i)  # Disposition des graphiques en 2 lignes et 3 colonnes
    indicator_data = cleaned_df[cleaned_df['Indicator Code'] == indicator]
    
    # Melt du DataFrame pour avoir 'Année' comme variable et 'Valeur' comme mesure
    melted_data = indicator_data.melt(id_vars=['Country Name', 'Indicator Name'], value_vars=years, var_name='Année', value_name='Valeur')
    
    sns.boxplot(x='Année', y='Valeur', data=melted_data)
    plt.title(indicator)
    plt.xticks(rotation=45)
    plt.yscale('log')  # Utiliser l'échelle logarithmique si nécessaire pour une meilleure lisibilité

plt.tight_layout()
plt.show()

```

```python
years = [str(year) for year in range(2000, 2016)]

# On crée un DataFrame contenant uniquement les données de ces années
data_for_years = cleaned_df[years]

# On initialise un dictionnaire pour stocker les indices des valeurs aberrantes pour chaque année
outliers_indices_dict = {}

# On va parcourir les années et identifier les valeurs aberrantes pour chaque année
for year in years:
    Q1 = data_for_years[year].quantile(0.25)
    Q3 = data_for_years[year].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (data_for_years[year] < lower_bound) | (data_for_years[year] > upper_bound)
    outliers_indices_dict[year] = data_for_years[outliers].index

# On va afficher les indices des valeurs aberrantes pour chaque année
for year, indices in outliers_indices_dict.items():
    if len(indices) > 0:
        print(f"Indices des valeurs aberrantes pour l'année {year} :")
        print(indices.tolist())
        print()
    else:
        print(f"Pas de valeurs aberrantes détectées pour l'année {year}.")
        print()
        
```

```python
for year in years:
    if len(outliers_indices_dict[year]) > 0:
        outlier_values = data_for_years.loc[outliers_indices_dict[year], year]
        min_outlier = outlier_values.min()
        max_outlier = outlier_values.max()
        print(f"Année {year}: Min des valeurs aberrantes = {min_outlier}, Max des valeurs aberrantes = {max_outlier}")
    else:
        print(f"Pas de valeurs aberrantes détectées pour l'année {year}.")
```
